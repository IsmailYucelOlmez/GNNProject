```python
from rdkit import Chem
from rdkit.Chem.Draw import MolToImage
import zipfile
import torch
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def mol_file_to_mol():

    pt_file = torch.load("./data/processed/data_0.pt")
    print(pt_file)

    
    if pt_file is not None:
         mol=Chem.MolFromSmiles(pt_file.smiles)
         svg=draw_molecule(mol)
    else:
        print("Invalid MOL file.")
    
    return svg

def draw_molecule(mol):
    
    return MolToImage(mol)
```
